```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples = 750, factor =0.3, noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
dbcan = DBSCAN(eps = 0.1)
```

```python
dbcan.fit_predict(X)
```

```python
dbcan.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=dbcan.labels_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get indices of outliers (noise points)
outlier_indices = np.where(dbcan.labels_ == -1)[0]

# Extract outlier points from your data X
outliers = X[outlier_indices]

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=dbcan.labels_, cmap='viridis', alpha=0.6)

# Overlay the outliers in red
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', label='Outliers', edgecolor='k')

plt.legend()
plt.show()
```
